# Study the effect of tunnel width on low-level in ST-RRT* and MILP

```python
import sys
sys.path.append('../../')
import numpy as np
from panav.environment.env import MultiTunnelEnv
from benchmarks.ST_RRTStar_wrapper import ST_RRTStar_wrapper, sequential_ST_RRTStar
from panav.SAMP.solvers import Tube_Planning
from panav.sequential import sequential_planning
import pickle as pkl


from time import time
%load_ext autoreload
%autoreload 2
```

```python
N = 2
limits = [(-20,20),(-20,20)]
n_tunnel = 1
bloating_r = 0.5
vmax = 1.0


widths = np.array([2.2,2.3,2.4,2.5,2.6,2.7,2.8,2.9,3.0,3.5,4.0]+list(range(5,11))) * bloating_r
```

```python

runtimes = {'MILP':[],"ST-RRT*":[]}
plans = {'MILP':[],"ST-RRT*":[]}
for w in widths:
    env = MultiTunnelEnv(n_tunnel = n_tunnel, 
                        tunnel_width = w,
                        limits=limits, 
                        N_agent = N)
    t0 = time()
    plan = sequential_planning(Tube_Planning,env,vmax,bloating_r)
    runtimes['MILP'].append(time()-t0)
    plans['MILP'].append(plan)

    t0 = time()
    plan = sequential_ST_RRTStar(env,vmax,bloating_r)
    runtimes['ST-RRT*'].append(time()-t0)
    plans['ST-RRT*'].append(plan)
```

```python

with open('./data/TunnelWidth.pkl','wb') as fp:
    pkl.dump(dict(runtimes=runtimes,plans = plans),fp)
```

```python
from matplotlib import pyplot as plt
with open('./data/TunnelWidth.pkl','rb') as fp:
    results = pkl.load(fp)
    runtimes = results['runtimes']
    plans = results['plans']

from matplotlib import rc 
font = {'size'   : 22,'weight':'bold'}
rc('font', **font)

color_dict = {'MILP':'blue','ST-RRT*':'green'}
runtimes['MILP'] = np.array(runtimes['MILP'])
runtimes['ST-RRT*'] = np.array(runtimes['ST-RRT*'])
idx_to_plot = [i for i,w in enumerate(widths) if w not in np.array([2.4,2.6,2.7,2.8,2.9,3.5])*bloating_r]
plt.plot(widths[idx_to_plot],runtimes['MILP'][idx_to_plot],label = 'MILP',marker='x',color = color_dict['MILP'], linewidth = 3 ,
                       markersize = 10)
plt.plot(widths[idx_to_plot],runtimes['ST-RRT*'][idx_to_plot],label = 'ST-RRT*',marker = '+',color = color_dict['ST-RRT*'], linewidth = 3 ,
                       markersize = 10)
plt.xlabel('Tunnel Width')
plt.ylabel('Runtime/s')
plt.legend()

plt.tight_layout()
plt.savefig('./simFigures/TunnelWidth.pdf')
```

```python
from matplotlib import pyplot as plt
from panav.viz import animation
from IPython.display import HTML, display
from matplotlib import rcParams
from panav.util import interpolate_positions

idx = 10
plans_to_viz = plans['MILP'][idx]
w = widths[idx]
print(w)

env = MultiTunnelEnv(n_tunnel = n_tunnel, 
                    tunnel_width = w,
                    limits=limits, 
                    N_agent = N)
exec_tau = 1.0

rcParams['animation.embed_limit'] = 2**128
dt = exec_tau 
agents = range(len(plans_to_viz))
pos_to_ani = []

for cp in plans_to_viz:
    _,xs = interpolate_positions(*cp,dt)
    pos_to_ani.append(xs)
    
anim = animation(env, pos_to_ani, bloating_r, dt = dt,no_legend=True,displayAgentID=False)
display(HTML(anim.to_jshtml()))
plt.close() 
```
