# Utilities to load and visualize the GPPC map and scenario files

```python
from matplotlib import pyplot as plt
from mapLoaders import load_map, load_scenario
from matplotlib.colors import ListedColormap
```

## Load and visualize the two-dimensional map

```python

map_file = 'arena.map'

prob_map = load_map(map_file)

scenario_file = 'arena.map.scen'

df = load_scenario(scenario_file)
prob_idx = 101

cmap = ListedColormap(['lightgreen','white'])

# plt.imshow(loader.map)
plt.imshow(prob_map,cmap=cmap)
start = df.loc[prob_idx,'start_x':'start_y']
goal = df.loc[prob_idx,'goal_x':'goal_y']
plt.scatter(start[0],start[1],marker='o',color ='blue' ,label = 'start')

plt.scatter(goal[0],goal[1],marker='x',color ='red' ,label = 'goal')

dummy_obs = plt.scatter(0,0,marker = 's',color = 'lightgreen',label='obstacles')

plt.legend()

dummy_obs.set_visible(False)

plt.title('Map:{}\n Planning Problem No.{}'.format(df.at[1,'map'],prob_idx))
plt.show()
```
